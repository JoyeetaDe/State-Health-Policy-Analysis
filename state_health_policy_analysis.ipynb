```python
## 5. Create Derived Variables
## Example: uninsured rate differences pre/post expansion
## Per capita calculations
## Binary indicators (expansion vs non-expansion)
```

```python
df_merged['pcp_to_hospital_ratio'] = df_merged['pcp_rate'] / df_merged['hospital_admissions_per_1000']
```

```python
df_merged['poverty_to_income_ratio'] = df_merged['pct_below_poverty_line'] / (df_merged['median_income'] / 100000)
```

```python
df_merged['adjusted_pcp_rate'] = df_merged['pcp_rate'] * (1 - df_merged['uninsured_rate'])
```

```python
df_merged['spending_per_life_year'] = df_merged['spending_2020'] / df_merged['life_expectancy']
```

```python
df_merged['edu_poverty_index'] = df_merged['bachelors_degree_or_higher'] - df_merged['pct_below_poverty_line']
```

```python
df_merged['income_per_poverty_point'] = df_merged['median_income'] / (df_merged['pct_below_poverty_line'] * 100)
```

```python
df_merged['income_per_edu'] = df_merged['median_income'] * df_merged['bachelors_degree_or_higher']
```

```python
df_merged['spending_intensity'] = df_merged['spending_2020'] / df_merged['pcp_rate']
```

```python
df_merged['expansion_spending_interaction'] = df_merged['spending_2020'] * df_merged['medicaid_expanded']
```

```python
df_merged.replace([float('inf'), -float('inf')], pd.NA, inplace=True)
print(df_merged.isna().sum())
```

```python
northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont', 'New Jersey', 'New York', 'Pennsylvania']
midwest = ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin', 'Iowa', 'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'South Dakota']
south = ['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia', 'District of Columbia', 'West Virginia',
         'Alabama', 'Kentucky', 'Mississippi', 'Tennessee', 'Arkansas', 'Louisiana', 'Oklahoma', 'Texas']
west = ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah', 'Wyoming',
        'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']

def assign_region(state):
    if state in northeast:
        return 'Northeast'
    elif state in midwest:
        return 'Midwest'
    elif state in south:
        return 'South'
    elif state in west:
        return 'West'
    else:
        return 'Other'

df_merged_cleaned['region'] = df_merged_cleaned['state'].apply(assign_region)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.stats.api as sms
from scipy.stats import shapiro
import numpy as np

# A helper function to run diagnostics for any OLS model
def run_basic_diagnostics(model, model_name):
    residuals = model.resid
    fitted = model.fittedvalues

    print(f"==============================")
    print(f"   Diagnostics for {model_name}")
    print(f"==============================\n")

    # ----------------------------
    # 1. Residuals vs Fitted Plot
    # ----------------------------
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=fitted, y=residuals, s=40)
    plt.axhline(0, color='red', linestyle='--')
    plt.xlabel("Fitted values")
    plt.ylabel("Residuals")
    plt.title(f"Residuals vs Fitted: {model_name}")
    plt.tight_layout()
    plt.show()

    # ----------------------------
    # 2. Histogram of Residuals
    # ----------------------------
    plt.figure(figsize=(6,4))
    sns.histplot(residuals, kde=True)
    plt.title(f"Residual Histogram: {model_name}")
    plt.xlabel("Residual")
    plt.tight_layout()
    plt.show()

    # ----------------------------
    # 3. Q–Q Plot
    # ----------------------------
    sm.qqplot(residuals, line='45', fit=True)
    plt.title(f"Q–Q Plot: {model_name}")
    plt.tight_layout()
    plt.show()

    # ----------------------------
    # 4. Breusch–Pagan Test
    # ----------------------------
    bp_test = sms.het_breuschpagan(residuals, model.model.exog)
    bp_labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']
    print("Breusch–Pagan Test:")
    for name, value in zip(bp_labels, bp_test):
        print(f"  {name}: {value}")
    print()

    # ----------------------------
    # 5. Shapiro–Wilk Normality Test
    # ----------------------------
    shapiro_stat, shapiro_p = shapiro(residuals)
    print("Shapiro–Wilk Test for Normality:")
    print(f"  Statistic: {shapiro_stat}")
    print(f"  p-value: {shapiro_p}")
    print()

    print("====================================\n\n")

# Run diagnostics for both models
run_basic_diagnostics(life_expectancy_model, "Life Expectancy Model")
run_basic_diagnostics(hospital_admissions_model, "Hospital Admissions Model")
```

```python
# Robust standard errors (HC3) for both models

# Life expectancy model with robust SEs
life_expectancy_model_robust = life_expectancy_model.get_robustcov_results(cov_type='HC3')
print("===== Life Expectancy Model (Robust SEs) =====")
print(life_expectancy_model_robust.summary())
print("\n")

# Hospital admissions model with robust SEs
hospital_admissions_model_robust = hospital_admissions_model.get_robustcov_results(cov_type='HC3')
print("===== Hospital Admissions Model (Robust SEs) =====")
print(hospital_admissions_model_robust.summary())
```
